```python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

# Load your .env file
load_dotenv()

# Configuration Constants
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
BASE_URL = "https://openrouter.ai/api" # The "Secret Sauce" for the Anthropic SDK
MODEL_NAME = "google/gemini-2.0-flash-001" # Fast and cheap for development

# Initialize the client once
client = Anthropic(
    base_url=BASE_URL,
    api_key=OPENROUTER_API_KEY
)
```

```python
messages = []

def add_user_message(messages_list, text):
    messages_list.append({"role": "user", "content": text})

add_user_message(messages, "What is the capital of Nepal? and some history")

# Using the stream context manager
with client.messages.stream(
    model=MODEL_NAME,
    messages=messages,
    max_tokens=1000,
    temperature=0.7
) as stream:
    # 'text_chunk' will be the actual string bits as they arrive
    for text_chunk in stream.text_stream:
        print(text_chunk, end="", flush=True)
```
