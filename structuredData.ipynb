````python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

# Load your .env file
load_dotenv()

# Configuration Constants
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
BASE_URL = "https://openrouter.ai/api" # The "Secret Sauce" for the Anthropic SDK
MODEL_NAME = "google/gemini-2.0-flash-001" # Fast and cheap for development

# Initialize the client once
client = Anthropic(
    base_url=BASE_URL,
    api_key=OPENROUTER_API_KEY
)
````

````python
messages = []

prompt = """Generate three different sample aws CLI commands. Each should be very short"""

def user_message(messages_list, content):
    messages_list.append({"role": "user", "content": content})
    
def add_assistant_message(messages_list, content):
    # This "primes" the model to start its response inside a code block
    messages_list.append({"role": "assistant", "content": content})
    
def chat(messages_list, stops=None):
    # FIXED: Added 'stops' parameter to the function definition
    response = client.messages.create(
        model=MODEL_NAME,
        max_tokens=1000,
        messages=messages_list,
        stop_sequences=stops  # Pass the stop sequences here
    )
    return response.content[0].text

# --- Execution ---

user_message(messages, prompt)

# Prime the assistant's response with backticks 
# Note: Use standard backticks ` rather than '
add_assistant_message(messages, "```bash")

# FIXED: Pass the stop sequence to the updated function
text = chat(messages, stops=["```"])

print(text)
````
