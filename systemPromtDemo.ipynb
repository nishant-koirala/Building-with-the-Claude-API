```python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

# Load your .env file
load_dotenv()

# Configuration Constants
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
BASE_URL = "https://openrouter.ai/api" # The "Secret Sauce" for the Anthropic SDK
MODEL_NAME = "google/gemini-2.0-flash-001" # Fast and cheap for development

# Initialize the client once
client = Anthropic(
    base_url=BASE_URL,
    api_key=OPENROUTER_API_KEY
)
```

```python
# Initialize as 'messages' (plural) to keep things clear
messages = []

# Update the function to accept 'text' as an argument
def add_user_message(messages_list, text):
    messages_list.append({"role": "user", "content": text})

# Update the function to handle the 'system' parameter
def chat(messages_list, system_prompt):
    response = client.messages.create(
        model=MODEL_NAME,
        max_tokens=1000,
        system=system_prompt, # Pass the system prompt here
        messages=messages_list,
    )
    return response.content[0].text

# --- Execution ---

# 1. Add the question to our list
add_user_message(messages, "What is the capital of Nepal and some history?")

# 2. Call the chat function with the system instructions
answer = chat(messages, system_prompt="Give a summary, in bullets being as much concise as possible")

print(answer)
```
